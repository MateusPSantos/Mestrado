```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tabela_resultados(url_file_mip,form_):
    
    df = pd.read_csv(url_file_mip,header=None, sep=';')

    table = pd.DataFrame()
    table = pd.concat([table, df], ignore_index=True)
    table.columns = ['instance','ub','lb','gap','time','nodes','opt']
    table = table[['instance','lb','ub','gap','time','nodes','opt']]

    table['grupo'] = 'grupo' 
   

    
    table_result = pd.DataFrame()

    contador  = 1
    set_cont = 1
    while True:
        set_ ='set0{}'.format(str(set_cont))
        tab = pd.DataFrame()
        for i in range(contador, contador+12):
            t = table.loc[table['instance'] == f'c52_{i}.txt']
            tab = pd.concat([tab,t])
            table.loc[table[table['instance'] == f'c52_{i}.txt'].index,'grupo'] = set_
            
        
        tab = pd.DataFrame({
            'instance':f'{set_}',
            'lb':tab['lb'].mean(),
            'ub':tab['ub'].mean(),
            #'gap':tab.loc[table['opt'] == 0,'gap'].mean(),
            #'time':tab.loc[table['opt'] == 1,'time'].mean(),
            'gap':tab['gap'].mean(),
            'time':tab['time'].mean(),
            'nodes':tab['nodes'].mean(),
            'opt':tab['opt'].sum(),
            },index=[f'{form_}']
        )


        if table_result.shape[0] == 0:
            table_result = tab.copy(deep=True)
        else:
            table_result = pd.concat([table_result,tab])

        contador = contador + 12
        set_cont = set_cont + 1

        if set_cont > 9 :
            break

    return table_result.reset_index(drop=True), table
```

```python
form_ = "mc"
mip_="mip"

file=f"C:/Users/mateu/Downloads/Mestrado-main/Mestrado-main/CLSP_PY/MODELO_MC/MODELO_MC_ORIGINAL/RESULTADOS/clsr_{form_}_{mip_}.txt"

table_result_mc, table_mc = tabela_resultados(file,form_)
```

```python
table_result_mc
```

```python
tab = pd.DataFrame({
    'instance':f'res_mip_mc',
    'lb':table_result_mc['lb'].mean(),
    'ub':table_result_mc['ub'].mean(),
    'gap':table_result_mc['gap'].mean(),
    'time':table_result_mc['time'].mean(),
    'nodes':table_result_mc['nodes'].mean(),
    'opt':table_result_mc['opt'].sum(),
    },index=[f'res_mip_mc']
)

res_mip_mc = pd.concat([table_result_mc, tab], ignore_index=True)
#res_mip_mc

print(res_mip_mc[['instance','lb','ub','gap','time','nodes','opt']].to_latex(index=False,float_format="%.2f"))
```
